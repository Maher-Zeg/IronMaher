```python
diamonds = pd.read_csv(r'C:\Users\Utilisateur\Desktop\Project 4\diamonds_clean.csv')
diamonds
```

```python
#linear
plt.figure(figsize=[10,10])
plt.plot(diamonds['price'], diamonds['carat'])
plt.xlabel('carat')
plt.ylabel('price')
plt.title("Price vs carat")
plt.legend()
plt.show()
```

```python
# bar
plt.figure(figsize=[10,10])
plt.bar(diamonds['clarity'], diamonds['carat'], label='bar')
plt.xlabel('clarity')
plt.ylabel('carat')
plt.title("Clarity vs Count")
plt.legend()
plt.show()
```

```python
#scatter
plt.figure(figsize=[10,10])
plt.scatter(diamonds['price'], diamonds['carat'], label = 'scatt')
plt.xlabel('carat')
plt.ylabel('price')
plt.title("Price vs carat")
plt.legend()
```

```python
# boxplot
plt.figure(figsize=[10,10])
plt.boxplot( x = 'carat', data=diamonds)
plt.xlabel('carat')
plt.title('Carat')
plt.legend()
```

```python
#Line
sns.lineplot(x='carat', y='x', data=diamonds, hue='cut')
plt.xlabel('Clarity')
plt.ylabel('Price')
plt.title('Bar Linear: Carat (Based on Cut)')
```

```python
# scatter
sns.scatterplot(x='carat', y='price', hue='cut', size='depth', data=diamonds)
plt.xlabel('Carat Weight')
plt.ylabel('Price')
plt.title('Scatter Plot: Carat Vs Price (Based on Cut)')
```

```python
#Barplot
fig, ba_axis = plt.subplots(1, 2, figsize=(12,8))
sns.barplot(x='color', y='price', data=diamonds, ax = ba_axis[0], color='salmon')
plt.xlabel('Color')
plt.ylabel('Price')
plt.title('Bar Plot: Color Vs Price (Based on Cut)')
sns.barplot(x='clarity', y='price', data=diamonds, ax = ba_axis[1], color='purple')
plt.xlabel('Clarity')
plt.ylabel('Price')
plt.title('Bar Plot: Clarity Vs Price (Based on Cut)')
```

```python
labels = diamonds.cut.unique().tolist()
sizes = diamonds.cut.value_counts().tolist()
fig = plt.figure(figsize = (10, 10))
colors = sns.color_palette('bright')[0:5]
plt.pie(sizes, labels=labels,colors = colors, autopct='%.0f%%')
plt.title("Percentage of cut Categories")
plt.show()
```
